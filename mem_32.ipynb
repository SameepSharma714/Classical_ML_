```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler

import joblib
```

```python
df=pd.read_csv('covid_toy.csv')
```

```python
df
```

```python
df.head()
```

```python
df['fever'].fillna(df['fever'].mean(),inplace=True)
```

```python
df['fever']=df['fever'].fillna(df['fever'].mean())
df['has_covid'] = df['has_covid'].map({'Yes': 1, 'No': 0})
df.head()
```

```python
df.head()

df.drop(['gender','cough','city'],axis=1,inplace=True)
X=df.drop('has_covid',axis=1)
y=df['has_covid']


```

```python
df.isnull().sum()
df.head()


df.isnull().sum()

```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
SR=StandardScaler()
```

```python
X_train_new=SR.fit_transform(X_train)
X_test_new=SR.fit_transform(X_test)


```

```python
lr=LinearRegression()
df.isna().sum().sum()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
plt.plot(X_train_new,lr.predict(X_train_new),color='r')
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)

```
